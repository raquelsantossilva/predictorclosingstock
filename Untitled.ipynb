```python
import pandas as pd
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
import joblib
```

```python
data =pd.DataFrame(pd.read_csv("googleprices.csv"))
```

```python
data #This dataset contains daily historical stock price data 
#for Google LLC (Ticker: GOOGL) covering the last 5 years.
#It includes essential financial metrics such as opening price, 
#daily high and low, closing price, adjusted close price, and trading volume.

```

```python
data.drop(0,inplace=True)
```

```python
for e in data.columns:
    if data[e].dtype =="object":
        try:
            data[e] =pd.to_numeric(data[e], errors='coerce')
        except:
            print("erro na conversÃ£o")
```

```python
#data.hist(bins=50,figsize=(20,15))
```

```python
x =data.iloc[:,[0, 2, 3, 4, 5]]
x
```

```python
y =data.iloc[:,1]
```

```python
x_treinamento,x_teste,y_treinamento,y_teste =train_test_split(x,y,test_size=0.2,random_state=0)
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import classification_report
```

```python
randomforestgoogle = RandomForestRegressor(random_state=42)
randomforestgoogle.fit(x_treinamento,y_treinamento)
```

```python
previsoes =randomforestgoogle.predict(x_teste)
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

MAE= mean_absolute_error(y_teste,previsoes)
MSE =mean_squared_error(y_teste,previsoes)
R2 = r2_score(y_teste,previsoes)

print(f"MAE: {MAE}\n")
print(f"MSE: {MSE}\n")
print(f"R2: {R2}\n")
```

```python
import matplotlib.pyplot  as plt

plt.scatter(y_teste,previsoes,alpha=0.6)
plt.show()
```
